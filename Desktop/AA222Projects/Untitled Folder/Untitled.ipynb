```python
import numpy as np
import matplotlib.pyplot as plt

radii = np.linspace(0.2,1,5)
thetas = np.linspace(0,2*np.pi,20)
theta, r = np.meshgrid(thetas, radii)

alpha = 2
ut = 0
ur = -0.5 * alpha * r

f = plt.figure()
ax = f.add_subplot(111, polar=True)
ax.quiver(theta, r, ur * np.cos(theta) - ut * np.sin(theta), ur * np.sin(theta))
```

```python
n = 7
r = np.linspace(0.1,1,n)
x = np.linspace(-1,1,n)
xx, rr = np.meshgrid(x,r)
thetas = np.linspace(0,2*np.pi,n)*np.ones((n,n))

alpha = 2
ux = alpha * xx
ur = -0.5 * alpha * rr
uy = ur * np.cos(thetas)
yy = rr * np.cos(thetas)

fig, ax = plt.subplots()
ax.quiver(xx, yy, ux, uy)
```
